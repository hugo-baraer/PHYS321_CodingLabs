## Question 5

The probability distribution for the speed $v$ of a particle in an ideal gas is given by the _Maxwell-Boltzmann_ distribution:

\begin{equation}
4 \pi \left(\frac{m}{2 \pi k_b T} \right)^{3/2}  v^2 \exp \left( \frac{-mv^2}{2 k_b T} \right),
\end{equation}

where $k_b$ is Boltzmann's constant, $m$ is the mass of an atom in the gas, and $T$ is the temperature. You can integrate this to confirm that it's correctly normalized if you like! (Incidentally, the wikipedia page on the Maxwell-Boltzmann distribution contains some typos, and a bunch of their expressions are incorrect).

__a)__ Suppose we have an ideal gas in the lab. Let's assume that we know what type of gas we have, so the mass is known. To simplify matters, work in units where $m / k_b = 1$. You go into the lab and you make _one_ measurement of velocity (i.e., you measure the speed of _one_ atom). You get $v = 20$ (in our weird units).

You'd like to know the temperature of the gas. Prof. Mxyptlk says "the temperature measures the _average_ kinetic energy of a particle; therefore, with just a single speed measurement we can't know anything about the temperature". Prove Prof. Mxyptlk wrong by inferring what you can about $T$, the temperature of the gas.

```python
#import the necessary packages
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

#to prove this Prof. off, let's plot the posterior distribution for Temperature


#let's assume a Gaussian prior distribution, with a mean at the normal lab temperature (293 kelvins) and a very large standard deviation.
# I could have picked a uniform prior distribution, but it's not very statistically significant.

#set an array of all the possible value of T
T_value = np.linspace(10,2000,100)
dxt = T_value[1]-T_value[0] #save the difference between those values as dx for future normalization. 

#set mean and standard deviation for prior
m = 293
std = 600

#saved the measured particle's velocity
v_measured = 20
#set the prior as a Gaussian and normalize it
prior_T = norm.pdf(T_value,m,std)
norm_prior_T = prior_T/(np.sum(prior_T*dxt))

#Using the tip that m/kb is = 1 in the Maxwell-Boltzmann distribution, plug it as the likelihood, and normalize it
likelihood_T = 4*np.pi*(1/(2*np.pi*T_value))**(3/2)*v_measured**2*np.exp(-v_measured**2/(2*T_value))
norm_likelihood_T= likelihood_T/(np.sum(likelihood_T*dxt))

#set the posterior as the prior time the likelihood, and normalize it
posterior_T = prior_T * likelihood_T
norm_posterior_T = posterior_T/(np.sum(posterior_T*dxt))

#plot the 3 distributions 
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xticks(np.linspace(10,2000,20))
ax.set_xlabel("Temperature [K]") 
ax.set_ylabel("Probability density")
ax.plot(T_value,norm_likelihood_T, label = 'likelihood',linestyle='--',lw=2)
ax.plot(T_value, norm_prior_T, label = 'prior',linestyle='-',lw=2)
ax.plot(T_value, norm_posterior_T, label = 'posterior',linestyle='-.',lw=2)
plt.legend()
plt.show()

```

```python
print("with this plot, we have an idea for what the temperature could be (red distribution on the graph). It's excpected mode value being", T_value[np.argmax(norm_posterior_T)])
```

__b)__ What is the 95\% credibility region for your inferred $T$?

```python
## This funtion was taken from lab 5


#Use the function created in coding Lab_5
def credible_region(posterior, b_frac, n_percentage):
    """
    Defines a region of a dsitribution where n% of the data is.
    :param posterior: [arr] the posterior distribution to analyse
    :param b_frac: [arr] an array of all the x values associated to the posterior distribution
    :param n_percentage: [float] the percentage of the credible region desire
    :returns: [arr] [int] [int] returns the n% credible region of the posterior distribution. It aslo returns the indeces of the boundary values in b_frac
    """
    #define the length of a segment for integration latr
    dx = b_frac[1]-b_frac[0]

    #define what spaces from 100% there should be on each side (to find the middle distribution we want)
    bound = (1.0 - n_percentage)/2
    #find the max and min percentage to look for in our dsitribtuion
    max_perc = 1.0 - bound
    min_perc = bound
    lower_b_val = 0 #set a value to 0, so later in the loop, we can take the colsest value to the boundary
    upper_b_val = 2000
    proba = [] #defines an empty list of probability
    
    #looping in the posterior distribtion
    for i in range(len(posterior)):
        proba.append(posterior[i]*dx) #add the probability of the posterior distribution to a ceratin point
        #when the probability of the distribution gets over our minimum value, and it's our first point to get over the limit
        if np.sum(proba) > (min_perc) and lower_b_val == 0: 
            lower_b_val = i #store that boundary b_frac to cut later
        #when the probability of the distribution gets over the maximum value
        elif np.sum(proba) > (max_perc): 
            upper_b_val = i #store that boundary b_frac to later
            break #get out of the loop
    
    #return the distribtion with cuts at the boundary value to keep only the central n% of the distribution
    #aslo returns the boundary value for the next question
    return posterior[lower_b_val:upper_b_val], lower_b_val, upper_b_val 

#define our desired 95% percentage
n_percentage = 0.95
#get the region from the above defiened function
t, lower_L_val, upper_L_val = credible_region(norm_posterior_T,T_value, n_percentage)


#plot this part of function
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xticks(np.linspace(10,2000,20))
ax.set_xlabel("Temperature [K]")
ax.set_ylabel("Probability density")
ax.plot(T_value[lower_L_val:upper_L_val],t, label = 'posterior',linestyle='-.',lw=2)
plt.title("The 95% credibility region of the temperature's posterior distribution")
plt.legend()
plt.show()

```

```python
# I used the retrograde thechnique, where the posterior distribution of the previous data point becomes the new prior distributions

#taking the function defined in lab 5 (modified), 

#Assumptions were made such as :
    
    # the first prior distribution for the temperature, as been defiened to be a Gaussian 
    # centered around the usual temperature for a lab and having a very high standard deviation
    
    #Temperature was assumed to go from 10 to 2000 kelvins
    
    #the likelihood function is assumed to be the  Maxwell-Boltzmann distribution,  with the assumption that m/kb = 1
    
    #posterior is assumed prior times likelihood

    #for the normlization, the difference between temperature values is assumed to be small enough to be like a dx in an integration
    
    #for the final answer of T, the excpected value is the mode, and the error is the values from which 68% is within


def plot_baesian_retro(v,T_value,m = 293.0, std = 600, my_prior = [],plot=True):
    """
    Plot the prior, likelihood ans posterior distributions of the temperature given the velocity of some molecules.
    :param v: [flt] the measured velocity
    :param T_value: [arr] an array of all the possible value of temperature 
    :param m: [float] the expected mean for our first prior
    :param std: [float] the expected standard deviation for our first prior
    :param my_prior: [list] a define prior that can be implemented as a past posterior. The dault is an empty list.
    :param plot: [bool] whether to plot of just return the data
    :return: [arrays] 3 1D arrays for the prior, likelihood and posterior distribution
    """
    #define dx and x as the difference between t values
    dxt = T_value[1]-T_value[0]
    #get the likelihood distribution with the function made for it
    likelihood_T = 4*np.pi*(1/(2*np.pi*T_value))**(3/2)*v_measured**2*np.exp(-v_measured**2/(2*T_value))
    norm_likelihood_T= likelihood_T/(np.sum(likelihood_T*dxt))

    #if the prior parameter is entered  
    if len(my_prior)> 0:
        #use it as the prior distribution
        norm_prior_T = my_prior
    #if it's the default parameter    
    else:
        prior_T = norm.pdf(T_value,m,std)
        norm_prior_T = prior_T/(np.sum(prior_T*dxt)) #use the defined prior function to get a pior based on theoretical assumptions

    posterior_T = norm_prior_T * norm_likelihood_T #use the defined posterior function to get a posterior based on you likelihood and you prior
    norm_posterior_T = posterior_T/(np.sum(posterior_T*dxt)) #normalize it

    if plot:
        #plot the 3 distributions, give them label and line style, to see if there is an overlay
        fig, ax = plt.subplots(figsize=(10,5))
        ax.set_xlabel("Temperature [K]")
        ax.set_ylabel("Probability density") 
        ax.plot(T_value,norm_likelihood_T, label = 'likelihood',linestyle='--',lw=2)
        ax.plot(T_value, norm_prior_T, label = 'prior',linestyle='-',lw=2) 
        ax.plot(T_value, norm_posterior_T, label = 'posterior',linestyle='-.',lw=2)
        plt.xticks(np.linspace(10,2000,20)) #set the x tick label
        plt.legend() 
        plt.show()

    return norm_likelihood_T, norm_prior_T, norm_posterior_T

#plot the value from assumptions
T_value = np.linspace(10,2000,100)
vs = [30,28.5,18,15,20,20] #recorded speed
post = [] #create an empty list for the first prior
for i in range(len(vs)): #for all velocity v
    l1, pr1, post = plot_baesian_retro(vs[i],T_value, my_prior=post) #plot a new posterior using the posterior from the previous data set as a new prior

    
#check the new info of T with the last posterior, like described in assumptions
t, lower_T_val, upper_L_val = credible_region(post,T_value, 68)
mode_T = T_value[np.argmax(post)]
error_T = (mode_T - T_value[lower_T_val])

print("The temperature is now assumed to be :", mode_T,"±",error_T, 'K')
```

__d)__ Prof. Mxyptlk says "ok, ok, but you had to know the mass of the gas. If you didn't know $m$ separately and had to infer $m$ and $T$ simultaneously, you wouldn't get a nice well-defined peak in your probability distribution for $T$. You'd basically find that any value of $T$ is possible." Do you agree with Prof. Mxyptlk?

* If you agree, explain why and give an example of a probability distribution where reasonable simultaneous constraints on $m$ and $T$ are still possible (i.e., imagine you have the ability to modify the laws of nature to slightly modify the Maxwell-Boltzmann distribution).
* If you disagree, explain why Prof. Mxyptlk is incorrect.


```python
## import modules
import numpy as np
import matplotlib.pyplot as plt

#for here, let's define our prior, such that both the temperature and the mass have uniform distributions : 
sidelength= 100

#this is from the search a rescue Python code (adapted) 

#defines temperature and mass from plausible value (uniform prior for them both)
temperatures = np.linspace(10, 20000, sidelength)
masses = np.linspace(0, 300, sidelength)

#merch them in a 2d grid
temperature_grid, masses_grid = np.meshgrid(temperatures, masses)

#define the speeds
v = [30,28.5,18,15,20,20]
num_pulses = len(v)


#create a posterior collection full of zero for the first iteration of the loop
posterior_collection = np.zeros((num_pulses, sidelength, sidelength))
#for all the speed
for i,x in enumerate(v):
    #the posterior is the uniform priors times the Maxwell Boltzman distribution, defiened as the likelihood function
    posterior_collection[i] = 4*np.pi*(masses_grid/(2*np.pi*temperature_grid))**(3/2)*v[i]**2*np.exp(-masses_grid *v[i]**2/(2*temperature_grid))
posterior_collection = np.cumprod(posterior_collection, axis=0)

#check the deltas temperature and masses for the normalization
delta_temperatures= temperatures[1] - temperatures[0]
delta_masses = masses[1] - masses[0]
area_element = delta_temperatures * delta_masses
for i in range(num_pulses): #for all the measured veolicites
    integral = np.sum(posterior_collection[i]) * area_element
    posterior_collection[i] /= integral
    #nomralize

    
#plot the obtain 2d grid with the color map
f, ax = plt.subplots(figsize=(8.5,6.5))
posterior_plot = ax.contourf(temperature_grid, masses_grid, posterior_collection[5], cmap='inferno')
ax.set_ylabel('mass [1/k_b]')
ax.set_xlabel('temperature [K]')
plt.colorbar(posterior_plot)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

#Plot the posteiror probability distribution of just the temperature (marginalize the masses)
temperature_pdf = np.sum(posterior_collection, axis=1) * delta_masses

f, ax = plt.subplots()

ax.plot(temperatures, temperature_pdf[5])
ax.set_ylabel('Probability')
ax.set_xlabel('temperature [K]')
plt.colorbar(posterior_plot)
plt.show()
```

## Question 6

Kevin likes dumplings. He goes to inspect a dumpling factory, where they claim that the length of their dumplings $L$ is Gaussian distributed with mean $4\,\textrm{cm}$ and a standard deviation of $1\,\textrm{cm}$. Kevin randomly selects $10$ dumplings and measures their lengths, obtaining

$L = (5.61, 2.22, 3.11, 6.01, 4.79, 3.70, 2.67, 3.90, 3.86, 3.48)\,\textrm{cm}$



__a)__ Use Bayesian inference to infer the posterior distribution for the mean and standard deviation of dumpling size at this factory. Display your results graphically.



where 

\begin{equation}
p(L\textrm{dumpling}) = \frac{1}{\sqrt{2 \pi \sigma_{prior}^2}} \exp \left[- \frac{(L_\textrm{dumpling} - L_\textrm{guess})^2}{2 \sigma_{prior}^2}\right],
\end{equation}
where $\sigma_p$ is the uncertainty in my prior knowledge.

as for the likelihood, 

5.61,2.22,3.11,6.01,4.79,3.70,2.67,3.90,3.86,3.48







add the std and mean in the plot

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

#taken and adjusted from lab 5
def plot_bayesian(L_measured,L_frac,m,std,plot=True):
    
    """
    Plot the prior, likelihood ans posterior distributions based on the length of Dumpling's sample.
    :param L_measured: [arr] a 1D array of the length of all the measured Dumpling
    :param L_frac: [arr] an array of all the possible length of Dumpling produced by the factory 
    :param m: [float] the expected mean for the length of Dumpling provided by the company 
    :param std: [float] the standard deviation for the length of Dumpling provided by the company 
    :param my_plot: [bool] whether to plot of just return the data
    :return: [array] 1D array of the posterior distribution
    """
    #define dx as the difference between the set L values, for normalization
    dx = L_frac[1]-L_frac[0]
    prior_dis = norm.pdf(L_frac,m,std)
    norm_prior_dis = prior_dis/(np.sum(prior_dis*dx))
    
    #likelihood 
    m_measured = np.mean(L_measured)
    std_measured = np.std(L_measured)
    likelihood_dis = norm.pdf(L_frac,m_measured,std_measured)
    norm_likelihood_dis = likelihood_dis/(np.sum(likelihood_dis*dx))
    
    #posterior
    posterior_dis = norm_likelihood_dis *  norm_prior_dis #get the posterior by multiplying the prior and likelihood distribution.
    norm_posterior_dis = posterior_dis/(np.sum(posterior_dis*dx))
    
    if plot:
        #plot the 3 distributions, give them label and line style, to see if there is an overlay
        fig, ax = plt.subplots(figsize=(10,5))
        ax.set_xticks(np.linspace(0,16,17))
        ax.set_xlabel("Length [cm]")
        ax.set_ylabel("Probability density")
        ax.plot(L_frac,norm_likelihood_dis, label = 'likelihood',linestyle='--',lw=2)
        ax.plot(L_frac, norm_prior_dis, label = 'prior',linestyle='-',lw=2)
        ax.plot(L_frac, norm_posterior_dis, label = 'posterior',linestyle='-.',lw=2)
        plt.legend()
        plt.show()


    return norm_posterior_dis


L_frac = np.linspace(0,15,1000)
L_measured = np.array([5.61,2.22,3.11,6.01,4.79,3.70,2.67,3.90,3.86,3.48])
m= 4.0
var = 1.0
posterior_value = plot_bayesian(L_measured,L_frac,m,var)


```
