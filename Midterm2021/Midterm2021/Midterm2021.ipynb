The probability distribution for the speed $v$ of a particle in an ideal gas is given by the _Maxwell-Boltzmann_ distribution:

\begin{equation}
4 \pi \left(\frac{m}{2 \pi k_b T} \right)^{3/2}  v^2 \exp \left( \frac{-mv^2}{2 k_b T} \right),
\end{equation}

where $k_b$ is Boltzmann's constant, $m$ is the mass of an atom in the gas, and $T$ is the temperature. You can integrate this to confirm that it's correctly normalized if you like! (Incidentally, the wikipedia page on the Maxwell-Boltzmann distribution contains some typos, and a bunch of their expressions are incorrect).

Prove Prof. Mxyptlk wrong by inferring what you can about $T$, the temperature of the gas.

__b)__ What is the 95\% credibility region for your inferred $T$?

```python
## Your answer here
def credible_region(posterior, b_frac, n_percentage):
    """
    Defines a region of a dsitribution where n% of the data is.
    :param posterior: [arr] the posterior distribution to analyse
    :param b_frac: [arr] an array of all the b values associated to the posterior distribution
    :param n_percentage: [float] the percentage of the credible region desire
    :returns: [arr] [int] [int] returns the n% credible region of the posterior distribution. It aslo returns the indeces of the boundary values in b_frac
    """
    #define the length of a segment for integration latr
    dx = b_frac[1]-b_frac[0]

    #define what spaces from 100% there should be on each side (to find the middle distribution we want)
    bound = (1.0 - n_percentage)/2
    #find the max and min percentage to look for in our dsitribtuion
    max_perc = 1.0 - bound
    min_perc = bound
    lower_b_val = 0 #set a value to 0, so later in the loop, we can take the colsest value to the boundary
    upper_b_val = 2000
    proba = [] #defines an empty list of probability
    
    #looping in the posterior distribtion
    for i in range(len(posterior)):
        proba.append(posterior[i]*dx) #add the probability of the posterior distribution to a ceratin point
        #when the probability of the distribution gets over our minimum value, and it's our first point to get over the limit
        if np.sum(proba) > (min_perc) and lower_b_val == 0: 
            lower_b_val = i #store that boundary b_frac to cut later
        #when the probability of the distribution gets over the maximum value
        elif np.sum(proba) > (max_perc): 
            upper_b_val = i #store that boundary b_frac to later
            break #get out of the loop
    
    #return the distribtion with cuts at the boundary value to keep only the central n% of the distribution
    #aslo returns the boundary value for the next question
    return posterior[lower_b_val:upper_b_val], lower_b_val, upper_b_val 

n_percentage = 0.95

t, lower_L_val, upper_L_val = credible_region(norm_posterior_T,T_value, n_percentage)
fig, ax = plt.subplots(figsize=(10,5))
ax.set_xticks(np.linspace(10,2000,20))
ax.set_xlabel("Temperature [u]") #where u is the unit of our weird temperature
ax.set_ylabel("Probability density")
ax.plot(T_value[lower_L_val:upper_L_val],t, label = 'posterior',linestyle='-.',lw=2)
plt.title("The 95% credibility region of the temperature's posterior distribution")
plt.legend()
plt.show()
```

$L = (5.61, 2.22, 3.11, 6.01, 4.79, 3.70, 2.67, 3.90, 3.86, 3.48)\,\textrm{cm}$



__a)__ Use Bayesian inference to infer the posterior distribution for the mean and standard deviation of dumpling size at this factory. Display your results graphically.



where 

\begin{equation}
p(mean_\textrm{dumpling}) = \frac{1}{\sqrt{2 \pi \sigma_{prior}^2}} \exp \left[- \frac{(L_\textrm{dumpling} - L_\textrm{guess})^2}{2 \sigma_{prior}^2}\right],

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

#taken and adjusted from lab 5
def plot_bayesian(L_measured,L_frac,m,std,plot=True):
    
    """
    Plot the prior, likelihood ans posterior distributions based on the length of Dumpling's sample.
    :param L_measured: [arr] a 1D array of the length of all the measured Dumpling
    :param L_frac: [arr] an array of all the possible length of Dumpling produced by the factory 
    :param m: [float] the expected mean for the length of Dumpling provided by the company 
    :param std: [float] the standard deviation for the length of Dumpling provided by the company 
    :param my_plot: [bool] whether to plot of just return the data
    :return: [array] 1D array of the posterior distribution
    """
    #define dx as the difference between the set L values, for normalization
    dx = L_frac[1]-L_frac[0]
    prior_dis = norm.pdf(L_frac,m,std)
    norm_prior_dis = prior_dis/(np.sum(prior_dis*dx))
    
    #likelihood 
    m_measured = np.mean(L_measured)
    std_measured = np.std(L_measured)
    likelihood_dis = norm.pdf(L_frac,m_measured,std_measured)
    norm_likelihood_dis = likelihood_dis/(np.sum(likelihood_dis*dx))
    
    #posterior
    posterior_dis = norm_likelihood_dis *  norm_prior_dis #get the posterior by multiplying the prior and likelihood distribution.
    norm_posterior_dis = posterior_dis/(np.sum(posterior_dis*dx))
    
    if plot:
        #plot the 3 distributions, give them label and line style, to see if there is an overlay
        fig, ax = plt.subplots(figsize=(10,5))
        ax.set_xticks(np.linspace(0,16,17))
        ax.set_xlabel("Length [cm]")
        ax.set_ylabel("Probability density")
        ax.plot(L_frac,norm_likelihood_dis, label = 'likelihood',linestyle='--',lw=2)
        ax.plot(L_frac, norm_prior_dis, label = 'prior',linestyle='-',lw=2)
        ax.plot(L_frac, norm_posterior_dis, label = 'posterior',linestyle='-.',lw=2)
        plt.legend()
        plt.show()


    return norm_posterior_dis


L_frac = np.linspace(0,15,1000)
L_measured = np.array([5.61,2.22,3.11,6.01,4.79,3.70,2.67,3.90,3.86,3.48])
m= 4.0
var = 1.0
posterior_value = plot_bayesian(L_measured,L_frac,m,var)


```
