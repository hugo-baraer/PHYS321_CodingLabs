In Coding Lab 01, we discussed how the sidereal day is shorter than a solar day by 4 minutes. Is it possible for the sidereal day on some planet to be __longer__ than the solar day? Explain. If it's helpful, you may include a diagram of some sort.

### My answer here

Yes, not only is it possible, but the planet right next to us, Venus, has a longer sideral day than it has a solar day (NSSDC – Venus Fact Sheet).  This is due to the rotational direction of the planet. For planet like Earth, where it's rotation is in the same direction as it's orbital motion, the sideral day (the time in between one full rotation) comes before the solar day (time in between the Sun is at a similar position in the sky). A diagram of the case is presented below :


## the sideral and solar days of Earth according to it's normal rotational direction

However, when the rotational direction is the opposite of the orbital motion direction (like in the case of Venus), than the sidereal day is longer than the solar day. This is call retrograde motion. In this case, the Sun will be overhead again before the planet can make a full rotation.
